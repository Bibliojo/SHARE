```python
urls = [
    'http://etd.aau.edu.et/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://export.arxiv.org/oai2?verb=ListRecords&metadataPrefix=oai_dc&from=2014-10-10',
    'http://repository.asu.edu/oai-pmh?verb=ListRecords&metadataPrefix=oai_dc&from=2014-10-05T00:00:00Z',
    'http://www.biodiversitylibrary.org/oai?verb=ListRecords&metadataPrefix=oai_dc&from=2015-02-01',
    'http://www.biomedcentral.com/search/results?terms=*&format=json&drpAddedInLast=1&itemsPerPage=250#2015-04-23',
    'http://scholarworks.boisestate.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'http://calhoun.nps.edu/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://digitalcommons.calpoly.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc&from=2014-10-05T00:00:00Z',
    'http://authors.library.caltech.edu/cgi/oai2',
    'https://www.repository.cam.ac.uk/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://digitalcommons.chapman.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'http://citeseerx.ist.psu.edu/oai2?verb=ListRecords&metadataPrefix=oai_dc',
    'http://ClinicalTrials.gov/show/NCT02425332?displayxml=true',
    'http://repository.cmu.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc&from=2014-10-14',
    'http://cogprints.org/cgi/oai2',
    'https://dspace.library.colostate.edu/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://academiccommons.columbia.edu/catalog/oai?verb=ListRecords&from=2014-10-01T00:00:00Z&until=2014-10-02T00:00:00Z&metadataPrefix=oai_dc',
    'http://api.crossref.org/v1/works?filter=from-pub-date:2015-02-02,until-pub-date:2015-02-02&rows=1000',
    'http://researchspace.csir.co.za/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://engagedscholarship.csuohio.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'http://academicworks.cuny.edu/do/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://scholar.colorado.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc&from=2015-04-22T00:00:00Z&to=2015-04-23',
    'http://cyberleninka.ru/oai?verb=ListRecords&metadataPrefix=oai_dc',
    'http://dailyssrn.com/rss/rss-all-2.0.xml',
    'http://dash.harvard.edu/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://oai.datacite.org/oai?verb=ListRecords&metadataPrefix=oai_dc',
    'https://cn.dataone.org/cn/v1/query/solr/?q=dateModified:[NOW-5DAY%20TO%20*]&rows=10',
    'http://dh.howard.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'http://www.osti.gov/pages/pagesxml?nrows=100&EntryDateFrom=01/01/2014&EntryDateTo=01/01/2015',
    'http://www.datadryad.org/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://dukespace.lib.duke.edu/dspace-oai/request',
    'https://api.github.com/repos/elifesciences/elife-articles/commits?since=2015-12-22&until=2015-12-31&page=1&per_page=100',
    'http://oai.erudit.org/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://api.figshare.com/v1/articles/search?search_for=*&from_date=2015-2-1&to_date=2015-2-1',
    'http://repository.lib.fit.edu/oai/request?verb=ListRecords',
    'https://biblio.ugent.be/oai',
    'http://bbytezarsivi.hacettepe.edu.tr/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'https://dataverse.harvard.edu/api/search?q=*&type=dataset&per_page=10&key=259d801c-6951-4841-bb32-5968faa41b45',
    'http://commons.lib.niu.edu/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://lib.dr.iastate.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'http://www.icpsr.umich.edu/icpsrweb/ICPSR/oai/studies?verb=ListRecords&metadataPrefix=oai_dc',
    'http://ir.uiowa.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'https://scholarworks.iu.edu/dspace-oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://digitalcommons.iwu.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'http://digitalcommons.kent.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'http://krex.k-state.edu/dspace-oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://researchonline.lshtm.ac.uk/cgi/oai2?verb=ListRecords&metadataPrefix=oai_dc',
    'http://130.179.67.140/api/3/action/current_package_list_with_resources',
    'http://mars.gmu.edu/oai/request',
    'http://darchive.mblwhoilibrary.org/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://dspace.mit.edu/oai/request?verb=ListRecords&metadataPrefix=oai_dc&from=2014-09-28',
    'https://mospace.umsystem.edu/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'https://commons.mla.org/deposits/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'http://www.nature.com/oai/request?verb=ListRecords&metadataPrefix=oai_dc&from=2016-03-05&until=2016-03-07"',
    'https://www.earthsystemgrid.org/oai/repository?verb=ListRecords&metadataPrefix=dif&from={}&until={}',
    'http://neurovault.org/api/collections/?format=json',
    'http://exporter.nih.gov/XMLData/final/RePORTER_PRJ_X_FY2015_035.zip',
    'https://materialsdata.nist.gov/dspace/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'https://dspace.nku.edu/oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://data.nodc.noaa.gov/cgi-bin/iso?id=gov.noaa.nodc:00001339;view=xml',
    'http://newprairiepress.org/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
    'http://api.nsf.gov/services/v1/awards.json',
    'http://oaktrust.library.tamu.edu/dspace-oai/request?verb=ListRecords&metadataPrefix=oai_dc',
    'http://opensiuc.lib.siu.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc&from=2014-10-09T00:00:00Z',
    'https://osf.io/api/v1/search/?q=category:registration%20AND%20NOT%20title=test%20AND%20NOT%20title=%22Test%20Project%22',
    'http://digitalcommons.pcom.edu/do/oai/?verb=ListRecords&metadataPrefix=oai_dc',
]

import requests
from gevent.pool import Pool
from furl import furl
import functools

def download_url(url):
    filename = './{}.txt'.format(furl(url).host)
    r = requests.get(url, stream=True)
    with open(filename, 'wb') as f:
        for chunk in r.iter_content(chunk_size=1024):
            if chunk:
                f.write(chunk)
    print(filename)

pool = Pool(20)
for url in urls:
    pool.spawn(functools.partial(download_url, url))

pool.join(timeout=45)
```
